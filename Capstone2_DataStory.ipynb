#### Determine the data points and features in the dataset
    Import the data from CSV file into Data frames

```python
import numpy as np
import pandas as pd
import json
from pandas.io.json import json_normalize
import seaborn as sns
import matplotlib.pyplot as plt

dictionary = pd.read_csv('data/MELBOURNE_HOUSE_PRICES_LESS.csv')
main       = pd.read_csv("data/Melbourne_housing_FULL.csv")
dict_df = pd.DataFrame.from_dict(dictionary, orient='columns', dtype=None)
main_df = pd.DataFrame.from_dict(main, orient='columns', dtype=None)
print(main_df.info())
print(main_df.head())
print(main_df.columns)
```

#### Influence of the feature "No. of Rooms" on Price

```python

# Personalities involved in various industries
_ = sns.lmplot(data=main_df,x='Rooms',y='Price',hue='CouncilArea',fit_reg=False,size=24, aspect=0.5)

plt.xticks(rotation=90)
_.set(ylim=(0,4000000))
plt.show()
```

Based on the above plot, 

1) Feature 'CouncilArea' is more influential than "Number of rooms" in determining the price of house.

2) The above claim can be further ascertained by the fact that there is a 12 bedroom house in the price range of 700K in 'Casey city council' whereas 2 bedroom house in Melbourne city council or Banyule city council costs the same.

#### Influence of Landsize and Type of the house on the price of the house.

```python
# Initialize Figure and Axes object
fig, ax = plt.subplots(figsize=(18,8))

_ = sns.barplot(x='Type',y='Price',hue='Regionname',data=main_df)
plt.show()
```

There are some interesting insights from the above plot

1) Price of house or Townhouse is higher in the Southern region than that of Eastern region. However, the price of a unit in the Eastern region is greater than its Southern counterpart.

2) Price of an unit is almost the same for all regions with the exception of Western Victoria. However, there are steep differences in the price of house.

3) Price of a house or a townhouse or an unit is almost the same in Northern Victoria i.e. there is no significant difference between the price of a house or town house or an unit.

#### Conclusion:

Based on the data visualizations, some interesting insights were derived and we understand that Location (Councilarea, Regionname) of the house play a key role in determining the price. Although several other features such as 'Number of Rooms', 'Car Spaces', 'Type of the house', 'Yearbuilt' does influence the price of a house, features such as 'Councilarea', 'Suburb' and 'Regionname' override the other factors.
